```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red


# Assuming you have a DataFrame named `launch_sites` and it contains a 'class' column

# Define a function to determine marker color based on the 'class' value
def determine_marker_color(row):
    if row['class'] == 1:
        return 'green'
    elif row['class'] == 0:
        return 'red'
    else:
        return 'black'  # You can choose another color for other cases if needed

# Apply the function to create the 'marker_color' column
spacex_df['marker_color'] = spacex_df.apply(determine_marker_color, axis=1)

# Display the DataFrame to check the new column
print(spacex_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
